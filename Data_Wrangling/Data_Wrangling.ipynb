#### COMMON DATA FORMATS

###### CSV ------------------------- XML --------------------------- JSON

#### Note: it is recommended to take the course Data Wrangling with mongoDB.

CSV: it has a series of rows. Each row corresponding to an entry. The first row of every file tells us what each entry means. The values of the first row can be called as the identifiers.

XML: the benefit in comparison to CSV is that it supports nested structures. 
A document element can have a series of tables, which can have a number of children (every row/column).

<Document Element>
    <Table>
        <Column1> Value <Column1/>
        <Column2> Value <Column2/>
        <Column3/> #Value missing               
    <Table/>
    
JSON: the benefit in comparison to CSV is that it supports nested structures. Json documents have a number of json objects. It looks like a python dictionary. We have keys, which corresponds to what we see in the header row in a CSV file, followed by values.

{
    'Key1': value1,
    'Key2': value2,
    'Key3': value3,
}

##### Note: data format is not a matter of the file extension, being .csv or .json. A file format really has to do with how the data is organized inside the file. We could have data that is formatted in json or csv, but that comes in a file whose extension is .txt

## WHICH METHODS CAN HELP TO DEAL WITH MISSING DATA?

#### PARTIAL DELETION: it means we limite the data we use for analysis to the data that we have available and we know is correct. It is divided into LISTWISE DELETION AND PAIRWISE DELETION.

#### LISTWISE DELETION means that the whole row is excluded for analysis.

#### PAIRWISE DELETION means that specific bad field of a row is excluded for analysis, but the good field is included for the analysis.




##### But sometimes we do not have enough data or there are scenarios where removing our missing values would compromisethe representativeness of our sample, it does not make sense to throw away a bunch of entries just because there are missing values. This could impact the statistical power of whatever analysis we try to do. In this case it makes sense to make an intelligent guess at the missing values in ou data. 

#### The process of approximating these missing values is called IMPUTATION.

There are different techniques of imputation. One of them is EASY IMPUTATION.

#### EASY IMPUTATION (CALCULATE MEAN OF THE COLUMN AND PUT IT IN THE CELLS WITH MISSING VALUES)
The good thing of this method is that we do not change the mean of the column. But if we are studying the relationship between the column where i applied EASY IMPUTATION  and another column, this method lessens the corelation between these two columns.

#### IMPUTE USING LINEAR REGRESSION
The general idea is that we create an equation that predicts missing values in our data, using imformation we do have, and then use that equation to fill in those missing values. The negative side effect of applying this technique is that we would over emphasize existing trends in the data. For example is there is a correlation between two columns, this method will amplify this trend. 

The two previous methods are the most simplistic ones. They are good but sometimes they can either attenaute or amplify preexisting trends in the data.

```python
list1=[]       
with open('turnstile-110528.txt', 'rb') as f:    
    reader=csv.reader(f)       
    for row in reader:
        for i in range(3,len(row),5):
            list1.append(row[0:3] + row[i:i+5])

with open("updated_turnstile-110528.txt", "wb") as f:
    writer = csv.writer(f)
    writer.writerows(list1)
```
